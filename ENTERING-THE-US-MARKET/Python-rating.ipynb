### Добавление столбцов с численностью населения, плотностью населения, площадью и медианным доходом домохозяйств по индексам

```python
population_list = []
land_area_in_sqkm_list = []
median_household_income_list = []
lng_list = []
lat_list = []
radius_in_m_list = []
#создание списков для записи значений

for code in rest_2019_grouped['postal_code']:
    
    try:
        search = SearchEngine()
        zipcode = search.by_zipcode(code)
        zipcode.zipcode

        population_list.append(zipcode.population)
        land_area_in_sqkm_list.append(int(zipcode.land_area_in_sqmi * 2.6))
        median_household_income_list.append(zipcode.median_household_income)
        lng_list.append(zipcode.lng)
        lat_list.append(zipcode.lat)
        radius_in_m_list.append(int(zipcode.radius_in_miles * 1609))
        #добавление значений в списки
   
    except:
        population_list.append(np.nan)
        land_area_in_sqkm_list.append(np.nan)
        median_household_income_list.append(np.nan)
        lng_list.append(np.nan)
        lat_list.append(np.nan)
        radius_in_m_list.append(np.nan)
        #заполнение nan в случае отсутствия данных
```

```python
population = pd.Series(population_list)
land_area_in_sqkm = pd.Series(land_area_in_sqkm_list)
median_household_income = pd.Series(median_household_income_list)
postal_code_lng = pd.Series(lng_list)
postal_code_lat = pd.Series(lat_list)
postal_code_radius_m = pd.Series(radius_in_m_list)
#перевод списков в тип Series для объединения с датафреймом
```

```python
postal_code_df = pd.concat([
    rest_2019_grouped,
    population,
    land_area_in_sqkm,
    median_household_income,
    postal_code_lng,
    postal_code_lat,
    postal_code_radius_m
], axis=1)
#объединение столбцов и датафрейма rest_2019_grouped
```

```python
postal_code_df = postal_code_df.rename(columns={0: 'population',
                                                1: 'land_area_in_sqkm',
                                                2: 'median_household_income',
                                                3: 'postal_code_lng',
                                                4: 'postal_code_lat',
                                                5: 'postal_code_radius_m'})
#переименование колонок
```

```python
postal_code_df['population_density'] = round(postal_code_df['population'] / postal_code_df['land_area_in_sqkm'], 2)
#расчёт плотности населения
```

## Расчёт рейтинга и определение порядка открытия ресторанов

```python
df_rest_2019_cities = postal_code_df
```

```python
df_rest_2019_cities['rating'] = (5 * df_rest_2019_cities['rest_number_per_sqkm']
                                 + 5 * df_rest_2019_cities['population_density']
                                 + 3 * df_rest_2019_cities['median_household_income']
                                 + 5 * df_rest_2019_cities['education_points'] 
                                 + 2 * df_rest_2019_cities['hospital_points'] 
                                 + 5 * df_rest_2019_cities['park_points'] 
                                 + 5 * df_rest_2019_cities['entertainment_points'] 
                                 + 5 * df_rest_2019_cities['stadium_points'] 
                                 + 5 * df_rest_2019_cities['accommodation_points'] 
                                 + 5 * df_rest_2019_cities['financial_points'])
#подсчёт рейтинга
```

```python
df_rest_2019_cities['normalized_rating'] = df_rest_2019_cities['rating'] / df_rest_2019_cities['rating'].max()
```

```python
df_rest_2019_sort = df_rest_2019_cities.sort_values(by=['normalized_rating'], ascending=False)
```
